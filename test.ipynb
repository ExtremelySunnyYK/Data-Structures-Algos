```python
import pandas as pd

df = pd.DataFrame(columns=[i for i in range(7)], index=[0,1,2,3,4,5,6])
p = 0.533

```

```python
df.loc[:,0] = 1
df.iloc[0] = 0
```

```python
df
```

```python
# F(n,m,p) = p *  P(F(n-1,m,p)  + (1-p) * P(F(n,m-1,p) )

for n in range(1,7):
    for m in range(1,7):
        df.iloc[m,n] = p * df.iloc[m,n-1] + (1-p) * df.iloc[m-1,n]

# df.iloc[1,1] = p * df.iloc[0,1] + (1-p) * df.iloc[1,0]
```
